```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Resources/Salary_Data.csv")
df.head()
```

```python
plt.scatter(df["YearsExperience"] , df["Salary"])
plt.xlabel("Years of Exp")
plt.ylabel("Salary")
plt.show()
```

```python
# number of rows, then columns, -1 means the number of rows is unspecified
X = df.YearsExperience.values.reshape(-1, 1)
#says that we have 30 rows and 1 column 
X.shape
```

```python
y = df["Salary"]
```

```python
#creating an instance of the model, instantiated! 
model = LinearRegression()
```

```python
#then learning stage --> fitting / training the model
model.fit(X, y)
```

```python
#after learning stage, use predict to generate predictions
y_pred = model.predict(X)
y_pred.shape
#gives back (30,) meaning 30 predictions
```

```python
plt.scatter(X,y)
plt.plot(X, y_pred, color="y")
plt.show()
```

```python
#so this model can extrapolate beyond the 10.5 years we have here:
model.coef_
#returns 9449
model.intercept_
#returns 25792
#so given y = mx + b we can say:
def hello(x):
    return (9449 * x) + 25792
```

```python
#which gives back the predicted salary for someone with 12 years exp
hello(12)
```
